```python
import pandas as pd
import numpy as np
import pickle
import re
from seq2seq.seq2seq_model import Embedding_Seq2SeqSummarizer
```

```python
model_dir_path = 'seq2seq/models'
data_dir_path = 'seq2seq/data'
```

```python
df = pd.read_csv(data_dir_path + "/lenta_test.csv")
X = df['text']
Y = df['title']
```

# Loading a model

```python
model_path = Embedding_Seq2SeqSummarizer.get_config_file_path(model_dir_path=model_dir_path)
with open(model_path, 'rb') as data:
    config = pickle.load(data)
    
summarizer = Embedding_Seq2SeqSummarizer(config)
summarizer.load_weights(weight_file_path=Embedding_Seq2SeqSummarizer.get_weight_file_path(model_dir_path=model_dir_path))
```

# Demo

```python
# Some examples that captured semantics
demo = [3, 5, 31, 36, 37, 47, 54, 55, 99, 19, 39, 119]
```

```python
# for i in range(10):
for i in demo:   
    x = X[i]
    actual_headline = Y[i]
    headline = summarizer.summarize(x)
    beam_headline = summarizer.beam_search(x)

    print(f'Article: {x}\n') 
    print(f'Original headline: {actual_headline}')        
    print(f'Greedy headline: {headline}')
    print(f'Beam search headline: {beam_headline}\n')
    print('\n\n')
```
